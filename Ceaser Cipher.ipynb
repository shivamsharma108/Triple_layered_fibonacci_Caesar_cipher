## FIBONACCI CAESAR CIPHER

```python
class FibonacciCaesar:
    def fibonacciTerm(self,n):
        if n==1 or n==2:
            return 1
        a=1
        b=1
        c=0
        i=3
        while i<=n:
            c=a+b
            a=b
            b=c
            i+=1
        return c
######################################################################################
    def encrypt(self,PT):
        CT=''
        PT_arr=PT.upper().split()
        for i in PT_arr:
            CT+=self.encrypt_helper(i)+' '
        return CT[:-1]

    def encrypt_helper(self,PT):
        n=len(PT)
        CT=''
        for i in range(n):
            ascii=ord(PT[i])
            CT+=chr(65+(ascii-65+self.fibonacciTerm(i+1))%26)
        return CT
######################################################################################

    def decrypt(self,CT):
        PT=''
        CT_arr=CT.upper().split()
        for i in CT_arr:
            PT+=self.decrypt_helper(i)+' '
        return PT[:-1]

    def decrypt_helper(self,CT):
        n=len(CT)
        PT=''
        for i in range(n):
            ascii=ord(CT[i])
            PT+=chr(65+(ascii-65-self.fibonacciTerm(i+1))%26)
        return PT
######################################################################################
#fibonacci Caesar Cipher:doesnt require a key




PT='YRVIM DRKPES'
obj=FibonacciCaesar()
CT=obj.encrypt(PT)
print(CT)
ans=obj.decrypt(CT)
print(ans)
```

## VIGNERE CIPHER

```python
class Vignere:
    def __init__(self):
        #Creating a vignere matrix
        self.vignere_mat=[]
        for i in range(26):
            t=[]
            for j in range(26):
                if i+j<26:
                    t.append(chr(i+j+65))
                else:
                    t.append(chr(i+j-26+65))

            self.vignere_mat.append(t)
            
#####################################################################################
    def encrypt_helper(self,PT,Key):
        CT=''
        n=len(PT)
        m=len(Key)
        #extending the key in case it is short
        while m<n:
            Key+=Key[:n-m]
            m=len(Key)
        else:
            Key=Key[:n]


        for i in range(n):
            col=ord(PT[i])-65
            row=ord(Key[i])-65
            CT+=self.vignere_mat[row][col]

        return CT
    
    def encrypt(self,PT,Key):
        CT=''
        PT_arr=PT.upper().split()
        Key=Key.upper()
        for i in PT_arr:
            CT+=self.encrypt_helper(i,Key)+' '
        return CT[:-1]
#####################################################################################


    def decrypt_helper(self,CT,Key):
        PT=''
        n=len(CT)
        m=len(Key)
        #extending the key in case it is short
        while m<n:
            Key+=Key[:n-m]
            m=len(Key)
        else:
            Key=Key[:n]

        for i in range(n):
            row=ord(Key[i])-65
            for j in range(26):
                if self.vignere_mat[row][j]==CT[i]:
                    col=j
                    break
            PT+=self.vignere_mat[0][col]


        return PT



    def decrypt(self,CT,Key):
        PT=''
        CT_arr=CT.upper().split()
        Key=Key.upper()
        for i in CT_arr:
            PT+=self.decrypt_helper(i,Key)+' '
        return PT[:-1]

PT='RACHNA'
Key='MRIIRS'
obj=Vignere()
CT=obj.encrypt(PT,Key)
print(CT)
ans=obj.decrypt(CT,Key)
print(ans)
```

```python
#playfair cipher-old
class Playfair():#a key is passed first
    def __init__(self,key):
        key=key.upper()
        self.playfair_mat=[[0 for i in range(5)] for j in range(5)]
        self.createPlayfairMatrix(key)


    def findEmpty(self):
        for i in range(5):
            for j in range(5):
                if self.playfair_mat[i][j]==0:#empty
                    return (i,j)
        return (-1,-1) #no empty place

    def createPlayfairMatrix(self,key):
        present={}
        for i in range(26):
            present[chr(i+65)]=False
        
        for i in key:
            if not present[i]:#letter hasn't come in matrix yet
                row,col=self.findEmpty()
                if row==-1 and col==-1:
                    break
                else:
                    if i=='J'or i=='I':
                        self.playfair_mat[row][col]='I'#j and i are same
                        present['I']=True
                        present['J']=True
                    else:
                        self.playfair_mat[row][col]=i
                        present[i]=True
        else:
            row,col=self.findEmpty()
            while row!=-1 and col!=-1:
                for i in range(26):
                    letter=chr(65+i)
                    if not present[letter]:
                        if letter=='I' or letter=='J':
                            self.playfair_mat[row][col]='I'
                            present['I']=True
                            present['J']=True
                        else:
                            self.playfair_mat[row][col]=letter
                            present[letter]=True
                        break

                row,col=self.findEmpty()
    def findLetterIndex(self,l):
        if l=='I' or l=='J':
            for i in range(5):
                for j in range(5):
                    if self.playfair_mat[i][j]=='I':
                        return i,j
        else:
            for i in range(5):
                for j in range(5):
                    if self.playfair_mat[i][j]==l:
                        return i,j
            
########################################################################################

    def encrypt(self,PT):
        CT=''
        PT_arr=PT.upper().split()
        for i in PT_arr:
            CT+=self.encrypt_helper(i)+' '
        return CT[:-1]

    def encrypt_helper(self,PT):
        #making a list by splitting PT into pairs
        pairlist=[]
        n=len(PT)
        i=0#iterator

        while i<n-1:
            if PT[i]!=PT[i+1]:
                pairlist.append(PT[i]+PT[i+1])
                i+=2
            else:
                #add a bogus letter
                pairlist.append(PT[i]+'X')
                i+=1
        else:
            if i<n:
                pairlist.append(PT[i]+'X')
                i+=1
        CT=''

        for i in pairlist:
            #finding row and column of letters in playfair_mat
            #for 0th letter
            row1,col1=self.findLetterIndex(i[0])
            row2,col2=self.findLetterIndex(i[1])
            if col1!=col2 and row1!=row2:
                CT+=self.playfair_mat[row1][col2]+self.playfair_mat[row2][col1]
            elif col1==col2:
                if row1==4:
                    CT+=self.playfair_mat[0][col2]+self.playfair_mat[row2+1][col1]
                elif row2==4:
                    CT+=self.playfair_mat[row1+1][col2]+self.playfair_mat[0][col1]
                else:
                    CT+=self.playfair_mat[row1+1][col2]+self.playfair_mat[row2+1][col1]
            elif row1==row2:
                    if col1==4:
                        CT+=self.playfair_mat[row1][0]+self.playfair_mat[row2][col2+1]
                    elif col2==4:
                        CT+=self.playfair_mat[row1][col1+1]+self.playfair_mat[row2][0]
                    else:
                        CT+=self.playfair_mat[row1][col1+1]+self.playfair_mat[row2][col2+1]
        return CT

    ##################################################################################
    def decrypt(self,CT):
        PT=''
        CT_arr=CT.upper().split()
        for i in CT_arr:
            PT+=self.decrypt_helper(i)+' '
        return PT[:-1]
    
    
    def decrypt_helper(self,CT):
        pairlist=[]
        n=len(CT)
        for i in range(0,n-1,2):
            pairlist.append(CT[i]+CT[i+1])
        PT=''
        for i in pairlist:
            #finding row and column of letters in playfair_mat
            #for 0th letter
            row1,col1=self.findLetterIndex(i[0])
            row2,col2=self.findLetterIndex(i[1])
            if col1!=col2 and row1!=row2:
                PT+=self.playfair_mat[row1][col2]+self.playfair_mat[row2][col1]
            elif col1==col2:
                if row1==0:
                    PT+=self.playfair_mat[4][col2]+self.playfair_mat[row2-1][col1]
                elif row2==0:
                    PT+=self.playfair_mat[row1-1][col2]+self.playfair_mat[4][col1]
                else:
                    PT+=self.playfair_mat[row1-1][col2]+self.playfair_mat[row2-1][col1]
            elif row1==row2:
                    if col1==0:
                        PT+=self.playfair_mat[row1][4]+self.playfair_mat[row2][col2-1]
                    elif col2==0:
                        PT+=self.playfair_mat[row1][col1-1]+self.playfair_mat[row2][4]
                    else:
                        PT+=self.playfair_mat[row1][col1-1]+self.playfair_mat[row2][col2-1]
        #removing the letter X from PT
        PT=PT.replace('X','')
        return PT
        
        
        

Key='mriu'
text='manav rachna international institute of research and studies'
obj=playfair(Key)
CT=obj.encrypt(text)
print(CT)
ans=obj.decrypt(CT)
print(ans)
```

```python
#FINAL old (Playfair-Caesar-Vignere)
#playfair with 5x5 key matrix
class MixCipher():
    def __init__(self,Key):#key is a word
        self.Key=Key
        self.obj1=FibonacciCaesar()
        self.obj2=Vignere()
        self.obj3=Playfair(self.Key)
    
    def encrypt(self,PT):
        #playfair
        ct1=self.obj3.encrypt(PT)
        #playfair to caesar
        ct2=self.obj1.encrypt(ct1)
        #playfair-caesar-vignere
        ct3=self.obj2.encrypt(ct2,self.Key)
        return ct3
    
    def decrypt(self,CT):
        #vignere to caesar-playfair
        pt1=self.obj2.decrypt(CT,self.Key)
        #caesar to playfair
        pt2=self.obj1.decrypt(pt1)
        #ans
        pt3=self.obj3.decrypt(pt2)
        return pt3
    
myObj=MixCipher('mriu')#passing the key
text='Manav Rachna International Institute of Research and Studies'
ct=myObj.encrypt(text)
print(ct)
ans=myObj.decrypt(ct)
print(ans)
```

```python
#playfair cipher-old
class Playfair():#a key is passed first
    def __init__(self,key):
        key=key.upper()
        self.playfair_mat=[[0 for i in range(5)] for j in range(5)]
        self.createPlayfairMatrix(key)


    def findEmpty(self):
        for i in range(5):
            for j in range(5):
                if self.playfair_mat[i][j]==0:#empty
                    return (i,j)
        return (-1,-1) #no empty place

    def createPlayfairMatrix(self,key):
        present={}
        for i in range(26):
            present[chr(i+65)]=False
        
        for i in key:
            if not present[i]:#letter hasn't come in matrix yet
                row,col=self.findEmpty()
                if row==-1 and col==-1:
                    break
                else:
                    if i=='J'or i=='I':
                        self.playfair_mat[row][col]='I'#j and i are same
                        present['I']=True
                        present['J']=True
                    else:
                        self.playfair_mat[row][col]=i
                        present[i]=True
        else:
            row,col=self.findEmpty()
            while row!=-1 and col!=-1:
                for i in range(26):
                    letter=chr(65+i)
                    if not present[letter]:
                        if letter=='I' or letter=='J':
                            self.playfair_mat[row][col]='I'
                            present['I']=True
                            present['J']=True
                        else:
                            self.playfair_mat[row][col]=letter
                            present[letter]=True
                        break

                row,col=self.findEmpty()
    def findLetterIndex(self,l):
        if l=='I' or l=='J':
            for i in range(5):
                for j in range(5):
                    if self.playfair_mat[i][j]=='I':
                        return i,j
        else:
            for i in range(5):
                for j in range(5):
                    if self.playfair_mat[i][j]==l:
                        return i,j
            
########################################################################################

    def encrypt(self,PT):
        CT=''
        PT_arr=PT.upper().split()
        for i in PT_arr:
            CT+=self.encrypt_helper(i)+' '
        return CT[:-1]

    def encrypt_helper(self,PT):
        #making a list by splitting PT into pairs
        pairlist=[]
        n=len(PT)
        i=0#iterator

        while i<n-1:
            if PT[i]!=PT[i+1]:
                pairlist.append(PT[i]+PT[i+1])
                i+=2
            else:
                #add a bogus letter
                pairlist.append(PT[i]+'X')
                i+=1
        else:
            if i<n:
                pairlist.append(PT[i]+'X')
                i+=1
        CT=''

        for i in pairlist:
            #finding row and column of letters in playfair_mat
            #for 0th letter
            row1,col1=self.findLetterIndex(i[0])
            row2,col2=self.findLetterIndex(i[1])
            if col1!=col2 and row1!=row2:
                CT+=self.playfair_mat[row1][col2]+self.playfair_mat[row2][col1]
            elif col1==col2:
                if row1==4:
                    CT+=self.playfair_mat[0][col2]+self.playfair_mat[row2+1][col1]
                elif row2==4:
                    CT+=self.playfair_mat[row1+1][col2]+self.playfair_mat[0][col1]
                else:
                    CT+=self.playfair_mat[row1+1][col2]+self.playfair_mat[row2+1][col1]
            elif row1==row2:
                    if col1==4:
                        CT+=self.playfair_mat[row1][0]+self.playfair_mat[row2][col2+1]
                    elif col2==4:
                        CT+=self.playfair_mat[row1][col1+1]+self.playfair_mat[row2][0]
                    else:
                        CT+=self.playfair_mat[row1][col1+1]+self.playfair_mat[row2][col2+1]
        return CT

    ##################################################################################
    def decrypt(self,CT):
        PT=''
        CT_arr=CT.upper().split()
        for i in CT_arr:
            PT+=self.decrypt_helper(i)+' '
        return PT[:-1]
    
    
    def decrypt_helper(self,CT):
        pairlist=[]
        n=len(CT)
        for i in range(0,n-1,2):
            pairlist.append(CT[i]+CT[i+1])
        PT=''
        for i in pairlist:
            #finding row and column of letters in playfair_mat
            #for 0th letter
            row1,col1=self.findLetterIndex(i[0])
            row2,col2=self.findLetterIndex(i[1])
            if col1!=col2 and row1!=row2:
                PT+=self.playfair_mat[row1][col2]+self.playfair_mat[row2][col1]
            elif col1==col2:
                if row1==0:
                    PT+=self.playfair_mat[4][col2]+self.playfair_mat[row2-1][col1]
                elif row2==0:
                    PT+=self.playfair_mat[row1-1][col2]+self.playfair_mat[4][col1]
                else:
                    PT+=self.playfair_mat[row1-1][col2]+self.playfair_mat[row2-1][col1]
            elif row1==row2:
                    if col1==0:
                        PT+=self.playfair_mat[row1][4]+self.playfair_mat[row2][col2-1]
                    elif col2==0:
                        PT+=self.playfair_mat[row1][col1-1]+self.playfair_mat[row2][4]
                    else:
                        PT+=self.playfair_mat[row1][col1-1]+self.playfair_mat[row2][col2-1]
        #removing the letter X from PT
        PT=PT.replace('X','')
        return PT
        
        
        

Key='jig'
text='HEXO'
obj=Playfair(Key)
CT=obj.encrypt(text)
print(CT)
ans=obj.decrypt(CT)
print(ans)
```
